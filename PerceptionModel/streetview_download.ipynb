```python
def download(url, name): 
    conn = urlopen(url) 
    outimg = conn.read() 
    data_img = cv2.imdecode(np.asarray(bytearray(outimg), dtype=np.uint8), 1) 
    conn.close()#这里一定要关闭，不然爬几次之后会报连接错误 
    cv2.imwrite(name, data_img) #
#    plt.imshow(data_img) 
    print('Pic Saved!') 
    return data_img
```

```python
fp = open("spider_bd_map.txt","r")
a=np.array(fp.readlines())
print(a)
```

```python
start_end_points = a
start_point = start_end_points[0] 
start_point = start_point.strip('\n') 
start_point = start_point.split(',') 
start_point_jin = start_point[0] 
start_point_wei = start_point[1] 

end_point = start_end_points[1] 
end_point = end_point.strip('\n') 
end_point = end_point.split(',') 
end_point_jin = end_point[0] 
end_point_wei = end_point[1]
print(start_point_wei)
```

```python
jins = np.arange(float(start_point_jin), float(end_point_jin), 0.0005)
points_num = len(jins)
weis = np.linspace(float(start_point_wei), float(end_point_wei), points_num,0.0005)
print(weis,jins)
```

```python

for iPoint in range(points_num): 
    jin = jins[iPoint] 
    for iPoint in range(points_num): 
        wei = weis[iPoint] #这里要注意下，对应的经纬度没有街景图的地方，输出的会是无效图片 
        pid=str(jin) + "_" + str(wei)
        img_name = "G:\\" + str(jin) + "_" + str(wei) +".jpg" 
        url = "http://api.map.baidu.com/panorama/v2?ak=keyY&width=512&height=256&location="+str(jin)+","+str(wei)+"&fov=90&heading=90&pitch=15" # 你的KEY" 
        outimg = download(url, img_name)
        data.append([pid,jin,wei])
        #print(data)
fp.close
import pandas as pd
data=np.array(data)
save = pd.DataFrame(data, columns = ['id', 'jin', 'wei']) 
save.to_csv('C:\\Users\\lenovo\\python\\project\\data_classify2.csv',index=False,header=False)
print("done")
```
